```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

kashti = sns.load_dataset("titanic")
p1 = sns.boxplot(x = "survived",
                 y = "age",
                 showmeans = True,
                 meanprops = {"marker" : "*",
                               "markersize" : "12",
                               "markeredgecolor" : "red"},
                 data = kashti)
plt.xlabel("How many are survived", size = 18, weight = 'bold')
plt.ylabel("Age (years)", size = 18,weight = 'bold')
plt.title("Box plot of kitny doobay kitny bachy", size = 19,weight = 'bold')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

kashti = sns.load_dataset('tips')
graph = sns.FacetGrid(kashti, col ="sex",  hue ="day")
graph.map(sns.regplot, "total_bill", "tip").add_legend()
plt.show()
```
