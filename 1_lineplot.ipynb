## adding limits

```python
import seaborn as sns
import matplotlib.pyplot as plt

# now load dataset

phool = sns.load_dataset("iris")
# print(phool)

# draw a line plot
sns.lineplot(x = "sepal_length", y = "sepal_width", data = phool)
plt.title("Line Graph")
plt.xlim(2)
plt.ylim(1)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# now load dataset

phool = sns.load_dataset("iris")
# print(phool)

# draw a line plot
sns.lineplot(x = "sepal_length", y = "sepal_width", data = phool)
plt.title("Line Graph")
# sns.set_style("dark")
sns.set_style(style = "dark")
sns.set_style(style = None, rc = None)
plt.show()
```

## size figure

```python
import seaborn as sns
import matplotlib.pyplot as plt

# now load dataset
 
phool = sns.load_dataset("iris")
# print(phool)
plt.figure(figsize = (10,10))
plt.title("Line Graph")
# draw a line plot
sns.lineplot(x = "sepal_length", y = "sepal_width", data = phool)

plt.show()
```
