```python
import seaborn as sns
import matplotlib.pyplot as plt

kashti = sns.load_dataset("titanic")
sns.set(style = 'dark')
print(kashti)
sns.boxplot()
sns.boxplot(x = 'class', y = 'fare', hue = 'sex', data = kashti)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tip = sns.load_dataset("tips")
print(tip)
sns.boxplot(x = tip['total_bill'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
tip = sns.load_dataset("tips")
print(tip)
sns.boxplot(y = tip['total_bill'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
tip = sns.load_dataset("tips")
sns.boxplot(x = "tip", y = "day", hue = "smoker", data = tip, dodge = False)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
tip = sns.load_dataset("tips")
sns.boxplot(x = "tip", y = "day", hue = "smoker", data = tip, dodge = True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
tip = sns.load_dataset("tips")
sns.boxplot(x = "tip", y = "day", hue = "smoker", data = tip, palette = "Set2", dodge = False)
plt.show()
```
